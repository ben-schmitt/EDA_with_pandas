Anscombe's quartet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in data
# this is from the following Wikipedia page https://en.wikipedia.org/wiki/Anscombe%27s_quartet

df = pd.read_csv('anscombe.csv')
```

```python
df.round(2)
```

```python
df.describe().round(3)
```

```python
stats_table = df.describe().round(3)
stats_table = stats_table.iloc[[1, 2]]
stats_table
```

```python

fig, axs, = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)

axs[0, 0].set_title("1")
axs[0, 0].scatter(df.x1, df.y1 )
#axs[0, 0].set_xlabel("x")
#axs[0, 0].set_ylabel("y")

axs[0, 1].set_title("2")
axs[0, 1].scatter(df.x2, df.y2 )

axs[1, 0].set_title("3")
axs[1, 0].scatter(df.x3, df.y3 )

axs[1, 1].set_title("4")
axs[1, 1].scatter(df.x4, df.y4 )

fig.tight_layout()
plt.show()
```
